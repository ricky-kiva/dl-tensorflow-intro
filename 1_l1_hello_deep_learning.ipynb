We will train a neural network equivalent to this:

```
def hw_function(x):
  y = (2 * x) - 1
  return y
```



````python
# import libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np

print(tf.__version__)
````

build Sequential model, specifications:
*   1 layer (1 Dense)
*   1 neuron (units)
*   1 feature for input data (input_shape)

````python
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])
])
````

**Loss function**: measures the guessed answer from fitting, and measures how well/badly it is

**Optimizer function**: used to make another guess. Try to minimize loss based on *loss function*'s result

````python
# compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')
````

Providing the data:

a sample of **6 X's** & **6 Y's**

note: **Numpy** is the de facto standard in declaring model's inputs & outputs

````python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
````

````python
# Train the model, with 200 epochs
model.fit(xs, ys, epochs=250)
````

````python
# make prediction
print(model.predict([10.0])) # 10 as value to predict
````
